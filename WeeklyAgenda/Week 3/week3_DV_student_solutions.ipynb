Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
df.plot(y=["income","expense"], kind="bar")
```

# 3. Stack the bars

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.bar(index, income,width=0.5,label='income')
ax.bar(index, expense,width=0.5,bottom=income,label='expense')
ax.set_xticklabels(index, rotation = 90)

ax.legend()

plt.show()
```

# 3. Rotate the labels and set figsize

```python
fig, ax = plt.subplots(figsize=(9,6))

ax.bar(index, income,width=0.5,label='income')
ax.bar(index, expense,width=0.5,bottom=income,label='expense')
ax.set_xticklabels(index, rotation = 0)
ax.legend()

plt.show()
```

# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]

g = df.plot.bar(figsize=(9,6), rot=0)
g.set_xticklabels(labels)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.03, p.get_height()+0.5))
```

# 7. Plot the avg tip by size  (lineplot)

```python
size_by_average = tips.groupby("size")["tip"].mean()
plt.plot(size_by_average.index, size_by_average.values)
plt.xlabel("size")
plt.show()
```

```python
liste=list(tips.groupby("size")["tip"].mean())
fig, axes = plt.subplots()
axes.plot(liste)
axes.set_xlabel("size")
plt.style.use("ggplot")
```

# 8.Set the linestyle as "--"

```python
size_by_average = tips.groupby("size")["tip"].mean()
plt.plot(size_by_average.index, size_by_average.values,ls="--",color="blue")
plt.xlabel("size")
plt.show()
```

# 10. Set an additional dimension using size column

```python

plt.scatter(tips.tip, tips.total_bill, s=10, alpha=0.8, c='blue', edgecolor='black', linewidth=1)
cbar = plt.colorbar()
cbar.set_label('Size (1-6)')
plt.tight_layout()
plt.show()
```
